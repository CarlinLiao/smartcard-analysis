```python
# load dates for to examine induced demand on

tue_before =  pd.read_csv("../nanjing-data/20150303.csv", header=None)
wed_before = pd.read_csv("../nanjing-data/20150304.csv", header=None)
thu_before = pd.read_csv("../nanjing-data/20150305.csv", header=None)
tue_after =  pd.read_csv("../nanjing-data/20150428.csv", header=None)
wed_after = pd.read_csv("../nanjing-data/20150429.csv", header=None)
thu_after = pd.read_csv("../nanjing-data/20150430.csv", header=None)
tempcol = pd.read_csv("../nanjing-data/header.csv")
tue_before.columns=tempcol.columns
wed_before.columns=tempcol.columns
thu_before.columns=tempcol.columns
tue_after.columns=tempcol.columns
wed_after.columns=tempcol.columns
thu_after.columns=tempcol.columns
```

```python
# concatenate after dates together
frames = [tue_after, wed_after, thu_after]
all_after = pd.concat(frames)
all_after
```

```python
# Find ids that were not in BEFORE (tues,wed,thu) but are in AFTER(all)
id_in_both = list(set(tue_before['Card_Id']).intersection(all_after['Card_Id']))
all_unique=all_after[~all_after['Card_Id'].isin(id_in_both)]
id_in_both = list(set(wed_before['Card_Id']).intersection(all_after['Card_Id']))
all_unique=all_unique[~all_unique['Card_Id'].isin(id_in_both)]
id_in_both = list(set(thu_before['Card_Id']).intersection(all_after['Card_Id']))
all_unique=all_unique[~all_unique['Card_Id'].isin(id_in_both)]
all_unique
```

```python
# Number of unique user IDs in all_after, new line only
orig_no_unique=len(all_after.where(all_after['Line_Id']==3).dropna(how="any").Card_Id.unique())
orig_no_unique
```

```python
all_after_newline=all_unique.where(all_unique['Line_Id']==3).dropna(how="any")
```

```python
# New line and ids in AFTER only 
all_after_newline
```

```python
no_newriders=len(all_after_newline.Card_Id.unique())
no_newriders
```

```python
orig_no_unique
```

```python
# Induced demand in terms of no of individuals
from __future__ import division
induced_dem=no_newriders/orig_no_unique*100
induced_dem
```

```python
# Induced demand in terms of no of trips
len(all_after_newline)/len(all_after.where(all_after['Line_Id']==3).dropna(how="any"))*100
```
